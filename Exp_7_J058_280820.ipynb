```
y= pd.DataFrame(wine['target'])
```

```
y= pd.get_dummies(y)
```

```
c4 = w.corr()
```

```
sb.heatmap(c4, annot= True, cmap= 'coolwarm', fmt= '.2g')
```

```
from sklearn.preprocessing import StandardScaler
```

```
sc= StandardScaler()
```

```
x= sc.fit(w);
```

```
print(w.iloc[:,1])
```

```
means= []
stds= []
for i in range(0, w.shape[1]):
  mew = np.mean(w.iloc[:,i])
  sd= np.std(w.iloc[:,i])
  means.append(mew)
  stds.append(sd)
vals= {'Means': means, 'Standard': stds}
```

```
vals
```

```
m1= vals['Means']
s1= vals['Standard']
for j in range(0, w.shape[1]):

  w.iloc[:,j] = (w.iloc[:,j]- m1[j])/ s1[j]
```

```
w.head()
```

```
from sklearn.model_selection import train_test_split
```

```
x_train, x_test, y_train, y_test = train_test_split(w, y, test_size= 0.2)
```

```
print('X_train shape',x_train.shape)
print('y_train shape',y_train.shape)
print('X_test shape',x_test.shape)
print('y_test shape',y_test.shape)
```

```
import keras 
from keras import models
from keras import layers
from keras.utils import to_categorical
```

```
y_train= to_categorical(y_train)
y_test= to_categorical(y_test)
```

```
nn4= models.Sequential()

nn4.add(layers.Dense(10, activation= 'tanh', input_dim= (13)))
nn4.add(layers.Dense(5, activation= 'relu'))
nn4.add(layers.Dense(3, activation= 'softmax'))

nn4.summary()
```

```
nn4.compile(optimizer= 'adam', loss= ['categorical_crossentropy'], metrics= 'accuracy')
```

```
hist4= nn4.fit(x_train, y_train, epochs= 200, validation_split= 0.25)
```

```
nn4.evaluate(x_test, y_test)
```

```
loss_train = hist4.history['loss']
loss_val = hist4.history['val_loss']
epochs = range(1,201)

plt.plot(epochs, loss_train, 'k', label='Training loss')
plt.plot(epochs, loss_val, 'r', label='validation loss')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```
loss_train = hist4.history['accuracy']
loss_val = hist4.history['val_accuracy']

epochs = range(1,201)

plt.plot(epochs, loss_train, 'k', label='Training accuracy')
plt.plot(epochs, loss_val, 'r', label='validation accuracy')

plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
