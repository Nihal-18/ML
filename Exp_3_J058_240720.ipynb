```
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```
years = range(1980,2014) # from 1980 to 2013 
india = df_can.loc['India',years]
india.head()
```

```
india.plot();
```

```
india.plot(
    kind='line',
    figsize=(10,10)
)

plt.title('Immigration of Indians to Canada from 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

plt.text(1994,22000,"Peak")
plt.show();
```

```
years = range(1980,2014)
indopak = df_can.loc[['India','Pakistan'],years]

indopak1 = indopak.transpose()

indopak1

indopak1.plot(
    kind='line',
    figsize=(10,10)
)

plt.title('Immigration to Canada from India and Pakistan 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.text(1994,22000,"Peak")
plt.show()
```

```
# top 5 countries

df_can.sort_values(by='Total', ascending=False, inplace=True, axis='index')

df_can.head()

dftop = df_can.head(5)

dftop

dftop = dftop[years]

dftop1 = dftop.transpose()


dftop1.plot(
    kind='line',
    figsize=(10,10)
)

plt.title('Immigration to Canada from Top Five na 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.text(1994,22000,"Peak")
plt.show()
```

```
# Bottom 5 countries

df_can.sort_values(by='Total', ascending=True, inplace=True, axis='index')

df_can.head()

dftop = df_can.head(5)

dftop

dftop = dftop[years]

dftop1 = dftop.transpose()


dftop1.plot(
    kind='line',
    figsize=(10,10)
)

plt.title('Immigration to Canada from Bottom Five nations 1980 to 2013')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```
